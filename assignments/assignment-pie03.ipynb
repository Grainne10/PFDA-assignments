#### Task - Read in the csv file from a URL and create a pie chart of peoples email domains


```python
import pandas as pd

# Import the CSV file from a URL

df =pd.read_csv('https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download')

# I viewed the data in excel, there is no empty cells under the email domain.

df.head()

```

```python
# Our goal is to show a pie chart of the different email addresses, I split out the email address.
def show_domain(Email): 
    return Email.split('@')[-1]

# I create a new column showing the domain only

df.loc[:, 'domain'] = df['Email'].apply(show_domain)

df.head()
```

```python
# I will import matplot lib to create the pie chart

import matplotlib.pyplot as plt
import seaborn as sns # to show a different colour

chart = df['domain'].value_counts()
colors = sns.color_palette("Set3", n_colors=len(chart))

total_count = chart.sum()
plt.figure(figsize=(8,8))
plt.pie(chart, labels=chart.index,  autopct=lambda p: f'{p:.1f}%\n({int(p * total_count / 100)})', 
        startangle=90, colors=colors, wedgeprops=dict(edgecolor='white'))  

plt.axis('equal')
plt.title('Pie Chart of E-mail Domains ')
plt.legend(chart.index, title="E-mail Domains", loc="lower left", bbox_to_anchor=(1, 1))
plt.show()  
```

 
##Conclusion

The pie chart data has 1000 lines with different email domains . From the pie chart, we can see that there is quite an even distibution of the different email domains. example.org and example.com are the closest in popularity, example.net is the least popular but not drastically lower in popularity.
