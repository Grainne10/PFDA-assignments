```python
# Read in the file as a csv file
FILENAME= "weatherreadings1.csv"


import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(FILENAME, parse_dates=['localReportEndDateTime'])

# Rename the columns for easier reading
df.rename(columns={'localReportEndDateTime': 'Local Date Time', 
                'dryBulbTemperature_Celsius': 'Dry Bulb Temp (Â°C)'}, inplace=True)

# Format the 'Local Date Time' column to a more readable format
df['Local Date Time'] = df['Local Date Time'].dt.strftime('%Y-%m-%d %H:%M')

# I reviewed the file in excel and dropped some columns that I thought were irrelevant and make the excel column easier to view

df.drop(['reportStartDateTime', 'reportEndDateTime', 'siteId', 'collectionName', 'isOfficial', 'isPublic', 'isLatestVersion', 'observationType', 'softwareType',  
        'isDcnn',  'id', 'externalSiteId', 'createdDateTime', 'version', 'externalId'], axis=1, inplace=True)


df.to_excel('weatherreadings1.xlsx', index=False)
```

```python
#Plot the data for Dry Bulb Temperature over time



# Plotting with a stem plot

plt.figure(figsize=(14, 8))
plt.stem(df['Local Date Time'], df['Dry Bulb Temp (Â°C)'], linefmt='b-', markerfmt='bo')
plt.title('Dry Bulb Temperature Over Time Stem Chart')
plt.xlabel('Date and Time')
plt.text(x=0.5, y=-5.0, s='Each stem represents a 15-minute interval', fontsize=10, ha='center')
plt.ylabel('Dry Bulb Temperature (Â°C)')
xticks = df['Local Date Time'][::4] 
plt.xticks(xticks, rotation=50, ha='right')


plt.annotate('Highest Temp', xy=(1, 16), xytext=(6, 16), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate('Lowest Temp', xy=(31, 9.5), xytext=(30, 11), arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.savefig('../img/Dry Bulb Temperature over 1 day.png')
plt.show()


```
